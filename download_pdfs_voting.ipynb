```python
# Function to generate a unique file name if the file already exists
def get_unique_file_name(file_path):
    base, extension = os.path.splitext(file_path)
    counter = 1
    while os.path.exists(file_path):
        # Append a counter to the base name (before the extension)
        file_path = f"{base}_{counter}{extension}"
        counter += 1
    return file_path

# Function to download a file with unique naming
def download_file(file_url, output_path):
    # Ensure the file path is unique if the file already exists
    unique_output_path = get_unique_file_name(output_path)
    
    response = requests.get(file_url)
    if response.status_code == 200:
        with open(unique_output_path, 'wb') as file:
            file.write(response.content)
        print(f"Downloaded: {unique_output_path}")
    else:
        print(f"Failed to download: {file_url}")

# Function to get folder name from URL
def get_folder_name_from_url(url):
    parsed_url = urlparse(url)
    folder_name = f"{parsed_url.netloc}{parsed_url.path.replace('/', '_')}"
    return folder_name.strip('_')

# Function to process a single URL
def process_url(url):
    folder_name = get_folder_name_from_url(url)
    output_dir = os.path.join(base_output_dir, folder_name)
    os.makedirs(output_dir, exist_ok=True)

    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find all downloadable document links
        for link in soup.find_all('a', class_='downloadable-file', href=True):
            href = link['href']
            file_url = urljoin(base_url, href)  # Convert relative URL to absolute URL
            file_name = href.split('/')[-1]  # Get the file name from the URL

            # Ensure the file name has a .pdf extension if missing
            if not file_name.endswith('.pdf'):
                file_name += '.pdf'
            
            output_path = os.path.join(output_dir, file_name)
            print(f"Processing file: {file_name}")
            download_file(file_url, output_path)
    else:
        print(f"Failed to fetch the webpage: {url}")

# Base URL of the site
base_url = "https://www.justice.gov"

# Directory to save downloaded PDFs
base_output_dir = "downloaded_pdfs"

# Iterate through the list of URLs
for url in urls_list:
    print(f"Processing: {url}")
    process_url(url)
```
